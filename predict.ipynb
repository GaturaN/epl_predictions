```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv", index_col=0)
```

```python
matches.head()
```

```python
matches.shape
```

```python
# Converting the 'date' column to datetime format
matches['date'] = pd.to_datetime(matches['date'])
```

```python
# Picking data from only two seasons i.e 2021/2022 & 2022/2023

end_date = pd.to_datetime('2023-09-29') #end of 2022 season
start_date = pd.to_datetime('2022-08-04')  #beginning of 2021 season


matches = matches[(matches['date'] < end_date) & (matches['date'] > start_date)]
```
